```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("movies.csv")
```

```python
df.tail()
```

```python
df.shape
```

```python
features=['genres','keywords','tagline','cast','director']
features
```

```python
for i in features:
    df[i]=df[i].fillna('')
```

```python
combine=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
combine
```

```python
vectors=TfidfVectorizer()
```

```python
feature_vectors=vectors.fit_transform(combine)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
movie_name=input("Enter your favourite movie name: ")
```

```python
all_titles=df['title'].tolist()
```

```python
find_close_match=difflib.get_close_matches(movie_name,all_titles)
find_close_match
```

```python
close_match=find_close_match[0]
close_match
```

```python
index=df[df.title==close_match]['index'].values[0]
print(index)
```

```python
similarity_score=list(enumerate(similarity[index]))
```

```python
len(similarity_score)
```

```python
sorted_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print("Movies you might also like: \n")
i=1
for movie in sorted_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<=10):
        print(i,'.',title_from_index)
        i+=1
```
