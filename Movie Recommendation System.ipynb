```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("movies.csv")
```

```python
df.tail()
```

```python
df.shape
```

```python
features=['genres','keywords','tagline','cast','director']
features
```

```python
for i in features:
    df[i]=df[i].fillna('')
```

```python
combine=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
combine
```

```python
vectors=TfidfVectorizer()
```

```python
feature_vectors=vectors.fit_transform(combine)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
movie_name=input("Enter your favourite movie name: ")
```

```python
all_titles=df['title'].tolist()
all_titles
```

```python
find_close_match=difflib.get_close_matches(movie_name,all_titles)
find_close_match
```

```python
close_match=find_close_match[0]
close_match
```

```python
index=df[df.title==close_match]['index'].values[0]
print(index)
```

```python
similarity_score=list(enumerate(similarity[index]))
print(similarity_score)
```
